````python
import pandas as pd

df = pd.read_csv("stock-list-full.csv")
````

````python
df.head()
````

````python
for index, row in df.iterrows():
      #print(row["stock_symbol"]+" "+row["company_search_name"])
        company_name = row["company_search_name"]
        additional_search = row["additional_search_term"]
        exclude_domains = "-site:www.moneycontrol.com -site:www.bseindia.com" 
        search_term = company_name + " " + additional_search + " " + exclude_domains
        results_per_page = 5
        num_count = 2
        search_results = googlesearch.search("investor relations quarterly financial "+search_term, tld="co.in", num=results_per_page, stop=num_count)
        print("Links for "+company_name+ " :")
        for result in search_results:
            print(result)
````
